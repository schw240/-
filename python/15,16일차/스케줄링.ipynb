```python
import schedule
import time
import datetime
```

```python
def job():
    print("i'm working")
    
def current_time():
    print(datetime.datetime.now())
```

```python
schedule.every().minutes.do(job)
schedule.every().seconds.do(current_time)
```

```python
while True:
    schedule.run_pending()
    time.sleep(1)
```
