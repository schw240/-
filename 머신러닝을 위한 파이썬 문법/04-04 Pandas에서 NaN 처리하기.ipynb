```python
#NaN 처리하기
# 1. 제거하기
df.dropna(axis=0) #assign을 하거나 inplace=True를 하면 적용이 됨
#안하면 적용은 안됨
```

```python
df
```

```python
# 평균으로 채우기

df.fillna(df.mean(axis=0)) #axis = 0 이므로 열(column) 으로 평균을 내겠다는 의미
#axis = 1이면 행으로 평균을 내겠다는 의미
```

```python
df.fillna(df.mean(axis=1))
```

```python
# 앞의값으로 채우기 , padding 방법, 뒤에값으로 채우기
```

```python
df.fillna(method='ffill')
```

```python
#pad는 ffill과 동일
df.fillna(method='pad')
```

```python
#맨 마지막의 경우 뒤에값이 없으므로 안채워짐
df.fillna(method='bfill')
```

```python
# df.fillna(df.mean(axis=0)) 이거랑 동일한 방식
df.where(pd.notnull(df), df.mean(), axis='columns')
#df.mean(axis=0,1) 을 사용하여 평균내는 방향도 정해줄 수 있다. 
```
