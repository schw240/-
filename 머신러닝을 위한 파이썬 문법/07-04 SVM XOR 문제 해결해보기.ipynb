```python
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# AND and OR 먼저 구분해보기
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = [0,0,0,1] #AND 조건
```

```python
X
```

```python
y
```

```python
LinearSVM = svm.LinearSVC()
```

```python
LinearSVM.fit(X,y)
```

```python
LinearSVM.predict([[0,1]])
```

```python
LinearSVM.predict([[1,1]])
```

```python
xx, yy= np.mgrid[0:1:0.1, 0:1:0.1]
grid = np.c_[xx.flatten(), yy.flatten()]
```

```python
pred = LinearSVM.predict(grid)
colors = ['red' if i == 0 else 'blue' for i in pred]
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(xx.flatten(), yy.flatten(), color=colors, alpha=0.2)
```

```python
xx, yy= np.mgrid[-1:2:0.04, -1:2:0.04]
grid = np.c_[xx.flatten(), yy.flatten()]
```

```python
pred = LinearSVM.predict(grid)
colors = ['red' if i == 0 else 'blue' for i in pred]
```

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(xx.flatten(), yy.flatten(), color=colors, alpha=0.2)
```

```python
xx, yy= np.mgrid[-1:2:0.1, -1:2:0.1]
grid = np.c_[xx.flatten(), yy.flatten()]
pred = LinearSVM.predict(grid)
colors = ['red' if i == 0 else 'blue' for i in pred]
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(xx.flatten(), yy.flatten(), color=colors, alpha=0.2)
```

```python
 # AND and OR 먼저 구분해보기
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = [0,1,1,1] #OR 조건
```

```python
LinearSVM.fit(X,y)
xx, yy= np.mgrid[-1:2:0.06, -1:2:0.06]
grid = np.c_[xx.flatten(), yy.flatten()]
pred = LinearSVM.predict(grid)
colors = ['red' if i == 0 else 'blue' for i in pred]
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(xx.flatten(), yy.flatten(), color=colors, alpha=0.2)
```

```python
# XOR 구분해보기
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = [0,1,1,0] #XOR 조건
```

```python
LinearSVM.fit(X,y)
xx, yy= np.mgrid[-1:2:0.06, -1:2:0.06]
grid = np.c_[xx.flatten(), yy.flatten()]
pred = LinearSVM.predict(grid)
colors = ['red' if i == 0 else 'blue' for i in pred]
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(xx.flatten(), yy.flatten(), color=colors, alpha=0.2)
```

```python
#이제 커널을 이용해서 XOR문제 풀어보기
#XOR 구분해보기
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = [0,1,1,0]
```

```python
svm = svm.SVC()
```

```python
svm.fit(X,y)
```

```python
print(svm.predict([[0,0]]))
print(svm.predict([[0,1]]))
print(svm.predict([[1,0]]))
print(svm.predict([[1,1]]))
```

```python
xx, yy= np.mgrid[0:1:0.02, 0:1:0.02]
grid = np.c_[xx.flatten(), yy.flatten()]
pred = svm.predict(grid)
colors = ['red' if i == 0 else 'blue' for i in pred]
plt.scatter(X[:,0], X[:,1], c=y, cmap='viridis')
plt.scatter(xx.flatten(), yy.flatten(), color=colors, alpha=0.2)
```
