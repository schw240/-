```python
x = np.arange(1,30)
y = x**2
y2 = x**2 + 5*x
y3 = x**2 + 10*x + 20
plt.plot(x,y, 'r--', y2, 'b--', y3, 'k--')
plt.show()
```

```python
x = np.linspace(-6,6)
y1, y2 = np.sin(x), np.cos(x)
plt.plot(x, y1, 'r--', linewidth=1, label='sin(x)')
plt.plot(x, y2, 'b--', linewidth=1, label='cos(x)' )
plt.title("Sine & Cosine Curve")
plt.legend()
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi*(15*np.random.rand(N))**2

plt.scatter(x,y, s=area, c=colors, alpha=1) #alpha = 투명도
plt.show()
```
