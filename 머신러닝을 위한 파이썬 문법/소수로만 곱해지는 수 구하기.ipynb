```python
#소수 구하는 함수
def is_prime_num(num):
    a = []
    i = 2
    if num != 1:
        while num > 2:
            if num%i == 0:
                a.append(i)
                num = num//i
            i += 1                
            if i > num:
                i = 2
                    
    if len(a) == 0:
        print('소수로만 곱하기 불가')
        
    return a
```

```python
ans = []
for i in range(10, 10000):
    start = time.time()
    ans.append(is_prime_num(i))
print(time.time()-start)
```

```python
from math import sqrt
#선생님 코드
def is_prime2(num):
    for j in range(2,int(sqrt(num))+1):
        if num%j == 0:
            return False
    return True

mul_prime = False
n = int(input('2보다 크고 10에 12승보다 작은 자연수 입력='))
cnt = 0

for i in range(2, int(sqrt(n))+1):
    if is_prime(i):
        m = n%i
        if m == 0 and is_prime2(n//i):
            cnt += 1
            #print('{} == > {}, {}'.format(n,i,n//i))
            mul_prime = True
if not mul_prime: print('처리불가')
```
