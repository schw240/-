```python
#강사님 코드
f = open('./실습데이터/회의실in2.txt', 'r')
n = int(f.readline())
meeting = []
for i in range(n):
    st, et = map(int, f.readline().split())
    meeting.append((st,et)) #튜플형식으로 저장
```

```python
meeting
```

```python
meeting.sort(key=lambda x : (x[1],x[0]))
cnt = 0
et = 0
for s,e in meeting:
    if s>= et:
        et = e
        cnt += 1
print('회의가 가능한 횟수=' , cnt)
```
