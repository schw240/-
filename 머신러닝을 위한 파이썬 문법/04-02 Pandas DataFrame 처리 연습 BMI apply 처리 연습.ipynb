```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
def name_gen():
    name_xx = ''
    for i in range(5):
        letter = chr(random.randint(97,122)) #아스키코드에서 문자 
        range
        name_xx = name_xx + letter
    return name_xx
```

```python
name_gen()
```

```python
names = [ name_gen() for i in range(10)]
weights = [ random.randint(45,101) for xx in names]
heights = [ random.randint(150, 190) for xx in names]
```

```python
BMI_table = pd.DataFrame({
    '이름' : names,
    '몸무게' : weights,
    '키' : heights
})
```

```python
BMI_table
```

```python
#BMI구하기
#몸무게 / 키(/100) ** 2
```

```python
BMI_table['BMI'] = BMI_table['몸무게'] / (BMI_table['키'] / 100) ** 2
```

```python
BMI_table
```

```python
def health_info(h_index):
    if h_index < 20:
        return '저체중'
    elif h_index >= 20 and h_index <= 24:
        return '정상'
    elif h_index >= 24 and h_index <= 29:
        return '과체중'
    else:
        return '비만'
```

```python
for xx in BMI_table['BMI']:
    print(xx, "======>", health_info(xx))
```

```python
BMI_table['health'] = 1 #외워두기 일종의 테크닉. 인덱스가 없기때문에 만들기 위한 방법
for idx, info in enumerate(BMI_table['BMI']):
    if health_info(idx) == '저체중':
        BMI_table.health[idx] = '저체중'
    elif health_info(idx) == '정상':
        BMI_table.health[idx] = '정상'
    elif health_info(idx) == '과체중': 
        BMI_table.health[idx] = '과체중'
    elif health_info(idx) == '비만': 
        BMI_table.health[idx] = '비만'
```
