```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2,2), rng.randn(2,200)).T
plt.scatter(X[:,0], X[:,1])
plt.axis('equal')
```

```python
X.T
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA
mypca = PCA(n_components=2)
mypca.fit(X)
```

```python
mypca.components_
```

```python
def draw_vector(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops = {'color':'red', 'arrowstyle':'simple', 'linewidth':2, 'shrinkA':0, 'shrinkB':0}
    ax.annotate('', v1, v0, arrowprops = arrowprops)
```

```python
plt.scatter(X[:,0], X[:,1], alpha = 0.5)
for length,vector in zip(mypca.explained_variance_, mypca.components_):
    v = vector*3*np.sqrt(length)
    draw_vector(mypca.mean_, mypca.mean_ + v)
plt.axis('equal')
```

```python
dimpca = PCA(n_components=1)
dimpca.fit(X)
X_pca = dimpca.transform(X)
```

```python
X.shape
```

```python
X_pca.shape
```

```python
X_new = dimpca.inverse_transform(X_pca)
```

```python
plt.scatter(X[:,0], X[:,1], alpha=0.2)
plt.scatter(X_new[:,0], X_new[:,1], alpha=0.8)
plt.axis('equal')
```
