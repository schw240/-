```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
import platform
sns.set(style='darkgrid')

if platform.system() == 'Windows':
# 윈도우인 경우
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:    
# Mac 인 경우
    rc('font', family='AppleGothic')
rc('font', family=font_name)
matplotlib.rcParams['axes.unicode_minus'] = False
# 마이너스 입력할때 필요
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
tips.sample(20)
```

```python
tips.describe()
```

```python
#multinomial로 linear regression을 하기위해서는 데이터들을 인코딩해주어야함
sns.relplot(x='total_bill', y='tip', data=tips, hue='time', row='sex', col='day')
#관계를 보여주는 plot이라해서 relationship plot => relplot
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', row='sex', col='day')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', hue='size')
```

```python
sns.relplot(x='total_bill', y='tip',data=tips,hue='time',palette=['b','r'],size='size',col='time')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin', split=True, hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', split=True, hue='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm', hue='sex')
```
