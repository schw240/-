```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
```

```python
plt.scatter(X[:,0],X[:,1], s=30)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
list(y_kmeans).count(0), list(y_kmeans).count(1), list(y_kmeans).count(2), list(y_kmeans).count(3)
```

```python
list(y_true).count(0), list(y_true).count(1), list(y_true).count(2), list(y_true).count(3)
```

```python
centers = kmeans.cluster_centers_
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=20, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='black', s=100, alpha=0.5)
```

```python
sns.scatterplot(X[:,0], X[:,1], hue=y_kmeans)
```

```python
from sklearn.datasets.samples_generator import make_moons
X,y = make_moons(200, noise=0.05, random_state=0)
plt.scatter(X[:,0], X[:,1], s=50)
```

```python
sns.scatterplot(X[:,0], X[:,1], hue=y)
```

```python
list(y).count(0)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
centers = kmeans.cluster_centers_
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=20, cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='black', s=100, alpha=0.5)
```

```python
sns.scatterplot(X[:,0], X[:,1], hue=y_kmeans)
```

```python
from sklearn.cluster import DBSCAN
#입실론은 밀도측정 반지름
DBS = DBSCAN(eps=0.3, min_samples=15)
```

```python
DBS.fit(X)
```

```python
y_prediction=DBS.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_prediction, s=20, cmap='viridis')
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier(n_estimators=10)
svm_clf = SVC(gamma='auto')
```

```python
log_clf.fit(X, y)
rnd_clf.fit(X,y)
svm_clf.fit(X,y)
```

```python
log_y_pred = log_clf.predict(X)
```

```python
log_clf.score(X,y)
```

```python
plt.scatter(X[:,0], X[:,1], c=log_y_pred, s=20, cmap='viridis')
```

```python
rnd_y_pred = rnd_clf.predict(X)
```

```python
rnd_clf.score(X,y)
```

```python
plt.scatter(X[:,0], X[:,1], c=rnd_y_pred, s=20, cmap='viridis')
```

```python
svm_y_pred = svm_clf.predict(X)
```

```python
svm_clf.score(X,y)
```

```python
plt.scatter(X[:,0], X[:,1], c=svm_y_pred, s=20, cmap='viridis')
```
